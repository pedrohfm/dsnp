```python
import numpy as np
import matplotlib.pyplot as plt
#  
fig, ax = plt.subplots(nrows=5,
                       ncols=2,
                       figsize=(20,20)
                       )

fig.set(tight_layout=True,
        dpi=300)

# 
ax[0,0].plot([1,0.8],[1,0.5])
ax[0,0].set_title("A");
ax[0,0].set_ylabel("y-axis A");
ax[0,0].set_xlabel("x-axis A");

# 
x = np.arange(1,10)

ax[0,1].plot(x**2);
ax[0,1].plot(x**2.25);
ax[0,1].plot(x**np.log(x));
ax[0,1].set_title("B");
ax[0,1].set_ylabel("Price")
ax[0,1].set_ylabel("Quantity")
ax[0,1].legend(['A','B','C'])
ax[0,1].grid('True')

# 
dates = ['2020', '2021', '2022', '2023', '2024']
units = [800, 650, 1200, 890, 1000]

ax[1,0].bar(dates,units,
            width=0.9,
            edgecolor='black',
            color='grey'
            );
ax[1,0].set(title="C",
            ylabel = "Frequency",
            xlabel = "Year",
            ylim = [0,max(units)+200],
            facecolor='beige',
            );
ax[1,0].spines[['top','right']].set_visible(False)
ax[1,0].legend(['Cars sold'],
               frameon=False,
               fontsize=8
               )

# 
y = np.random.randint(low=0,high=1000,size=100)
z = np.random.randint(low=y-250,high=y,size=100)
mean = np.average((y+z)/2)

ax[1,1].set_title("D");
ax[1,1].scatter(x=y,y=z,label='Scores');
ax[1,1].legend(loc='best')
ax[1,1].axhline(mean,
                color='red',
                linewidth=0.7,
                linestyle='dashed'
                );
ax[1,1].annotate(f"Mean scores: {mean:.2f}",
                 xy=(0,mean),
                 xytext=(0,mean+20),
                 color='red',
                 fontsize=12,
                 );
# 
ax[2,0].set_title("E");

# 
ax[2,1].set_title("F");

# 
ax[3,0].set_title("G");

# 
ax[3,1].set_title("H");

# 
ax[4,0].set_title("I");

# 
ax[4,1].set_title("J");
```
