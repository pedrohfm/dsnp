```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=5,
                       ncols=2,
                       figsize=(10,10), 
                       dpi=300 #300 dpi is print quality
                       )

fig.tight_layout(pad=4);

ax[0,0].plot([1,0.8],[1,0.5])
ax[0,0].set_title("A");
ax[0,0].set_ylabel("y-axis A");
ax[0,0].set_xlabel("x-axis A");

x = np.arange(1,100)
ax[0,1].plot(x,x**2);
ax[0,1].set_title("B");

ax[1,0].set_title("C");
ax[1,0].set_("C");

ax[1,1].set_title("D");
ax[2,0].set_title("E");
ax[2,1].set_title("F");
ax[3,0].set_title("G");
ax[3,1].set_title("H");
ax[4,0].set_title("I");
ax[4,1].set_title("J");
```
