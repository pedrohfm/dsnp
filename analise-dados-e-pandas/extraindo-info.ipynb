# Getting information from data

Focus on descriptive statistics and some early stage exploration. 

```python
import pandas as pd

df = pd.read_csv('./df_rides.csv')
df
```

```python
df.describe()
```

```python
# Calculating the average ride duration
avg_ride_duration = round(df.ride_duration.mean(),2) 
print(f"Average ride duration is: {avg_ride_duration} minutes.")
```

```python
# We can remove the invalid entries.
df.ride_late.dropna()

df.ride_late.sum()

# Counting the values 0 - for not late and 1 - for late.
df.ride_late.value_counts()
# If we pass 1 to value_counts() we get the proportion.
df.ride_late.value_counts(1)

# Calculating percentage of people that were late to delivering the bike.
ride_late_proportion= round(df.ride_late.value_counts(1),4) 

# Now we have a list containing the proportions between the two values, so we can calculate the percentage.
ride_late_percentage = ride_late_proportion[0]*100
print(f"{ride_late_percentage}% of people delivered their bike late.")
```
